```python
from matplotlib import pyplot as plt
plt.rcParams["figure.figsize"] = [12, 8]
plt.style.use("dark_background")
```

```python
from genpg.dataset import read_test_data

(labels, images) = read_test_data()

n = 8
fig, axs = plt.subplots(n, 10)
for row in range(10):
    for col, i in enumerate([i for i, L in enumerate(labels) if L == row]):
        if col >= n:
            break
        ax = axs[col, row]
        ax.axis('off')
        ax.imshow(images[i], interpolation='nearest')
fig.show()
```
